```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

#### Задание 1: Обработка изображений
Используйте библиотеку `Pillow` или `opencv-python` для 
загрузки изображения в форматах jpg и png. Если загружен 
файл png, проверьте, содержит ли он прозрачные пиксели 
и если да, то сделайте такие пиксели белыми 
непрозрачными.

```python
image1 = cv2.imread('325377_960x640.jpg')
plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))
```

```python
image2 = cv2.imread('325377_960x640_processed.png', cv2.IMREAD_UNCHANGED)
plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
```

```python
# Проверка на наличие альфа-канала
if image2.shape[2] == 4:  # 4 канала: BGR + Alpha
    alpha_channel = image2[:, :, 3]
    # Находим прозрачные пиксели (альфа = 0)
    transparent_pixels = np.where(alpha_channel == 0)

    # Заменяем прозрачные пиксели на белые во всех каналах BGR
    image2[transparent_pixels[0], transparent_pixels[1], 0] = 255
    image2[transparent_pixels[0], transparent_pixels[1], 1] = 255
    image2[transparent_pixels[0], transparent_pixels[1], 2] = 255

    # Сохраняем измененное изображение (важно сохранить с альфа-каналом)
    cv2.imwrite("output.png", image2)
    print("Прозрачные пиксели успешно заменены на белые. Сохранено как output.png")
else:
    print("Изображение не содержит альфа-канала (прозрачности).")
```

```python
output = cv2.imread('output.png', cv2.IMREAD_UNCHANGED)
plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
```
